```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import os
```

```python
os.listdir('images')
```

```python
label = {
    'notsmoking': 0,
    'smoking': 1
}
```

```python
X_train = []
y_train = []
for i in os.listdir('images/Training/Training'):
    image = cv2.imread(f'images/Training/Training/{i}');
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image,(100,100))
    image = image / 255
    image = image.reshape(100,100,1)
    X_train.append(image)
    y_train.append(label[i[:i.find("_")]])

X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_test = []
y_test = []
for i in os.listdir('images/Testing/Testing'):
    image = cv2.imread(f'images/Testing/Testing/{i}');
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image,(100,100))
    image = image / 255
    image = image.reshape(100,100,1)
    X_test.append(image)
    y_test.append(label[i[:i.find("_")]])
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
from sklearn.utils import shuffle
```

```python
X_train, y_train = shuffle(X_train, y_train)
X_test, y_test = shuffle(X_test, y_test)
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train,num_classes=2)
y_test = to_categorical(y_test,num_classes=2)
```

```python
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
from keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=64, kernel_size=(5, 5), padding='same', activation='relu', input_shape=(100, 100, 1)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units=16,activation='relu'))
model.add(Dense(units=8,activation='relu'))
model.add(Dense(units=4,activation='relu'))
model.add(Dense(units=2,activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
print(len(X_train))
print(len(y_train))
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=30)
```
